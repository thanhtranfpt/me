```python
# import statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# XỬ LÝ FILE CSV BỊ LỖI:
def write_to_csv(path,one_row,method = 'a'):
    # one_row = ['c','de','guii','hij']
    import csv
    with open(path, method, encoding='UTF-8', newline='') as file_handle: #------------ EDIT TYPE OF WRITING: w = write, a = append -----------
        writer = csv.writer(file_handle)
        # write a row to the csv file #------ MAIN ---------
        writer.writerow(one_row)
def read_rows_csv(path):
    import csv
    lines = []
    with open(path) as file_handle:
        file_lines = csv.reader(file_handle)
        for line in file_lines:
            lines.append(line)
    return lines #------------ lines = [ ['a','b','c', ...],[...], ... ] each element is a row.

folder = r'K:\LEARN AI FPT\TERM 3\AI - DS with Python - SQL\Datasets\bank\\'
file = folder + 'bank.txt'
f = open(file,'r')
f = f.readlines()
for k in f:
    row = [dt.replace('"','').strip() for dt in k.strip().split(';')]
    '''if row[-1].strip() == 'no':
        row[-1] = 0
    elif row[-1].strip() == 'yes':
        row[-1] = 1'''
    write_to_csv(folder + 'bankk.csv',row)
```

```python
#drop columns which are not needed.
# KEEP: job   marital  default  housing  loan  poutcome    y = [1 2 4 6 7 15 16 ]
keep_cols = [1,2,4,6,7,15,16]
drop_cols = []
print(df.columns[9])
for k in range(len(df.columns)):
    if k not in keep_cols:
        drop_cols.append(k)
        
#drop columns which are not needed.
df.drop(df.columns[drop_cols], axis = 1, inplace = True)
```

```python
# creating one hot encoding of the categorical columns.
data = pd.get_dummies(df, columns =['job', 'marital', 'default', 'housing', 'loan', 'poutcome'])
```

```python
# drop columns that has no values (_unknown)

keep_cols = ['y', 'job_admin.', 'job_blue-collar', 'job_entrepreneur', 
'job_housemaid', 'job_management', 'job_retired', 'job_self-employed', 
'job_services', 'job_student', 'job_technician', 'job_unemployed',
'marital_divorced', 'marital_married', 'marital_single', 'default_no', 
'default_yes', 'housing_no', 'housing_yes', 'loan_no', 'loan_yes',
'poutcome_failure', 'poutcome_nonexistent', 'poutcome_success']

drop_cols = []
for k in range(len(data.columns)):
    if data.columns[k] not in keep_cols:
        drop_cols.append(k)

data.drop(data.columns[drop_cols], axis=1, inplace=True)
```

```python
X = data.iloc[:,1:]
```

```python
Y = data.iloc[:,0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, train_size=0.6)
```

```python
Y_test.head()
```

```python
classifier = LogisticRegression(solver='lbfgs',random_state=0,max_iter=1000000)
```

```python
classifier.fit(X_train, Y_train)
```

```python
predicted_y = classifier.predict(X_train[:1]) #-------here-----
```

```python
for x in range(len(predicted_y)):
    if (predicted_y[x] == 'yes'):
        print(x, end="\t")
```

```python
print('Accuracy: {:.2f}'.format(classifier.score(X_test, Y_test)))
```
